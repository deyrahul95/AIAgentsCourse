# Agentic AI 

```python
from dotenv import load_dotenv

load_dotenv(override=True)
```

```python
import os

openrouter_api_key = os.getenv("OPENROUTER_API_KEY")

if openrouter_api_key:
    print(f"Open Router API Key exists and begins with {openrouter_api_key[:8]}")
else:
    print("Please set the Open Router API Key into env variable")
```

```python
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
openai = OpenAI(
    base_url=os.getenv("OPENROUTER_API_URL"), api_key=os.getenv("OPENROUTER_API_KEY")
)
```

```python
response = openai.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=[{"role": "user", "content": "What is 2 + 2?"}],
)
print(response.choices[0].message.content)
```

```python
response = openai.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=[
        {
            "role": "user",
            "content": "Could you please help me to pick a business area that might be worth exploring for an Agentic AI opportunity?",
        }
    ],
)
business_area = response.choices[0].message.content
display(Markdown(business_area))
```

```python
response = openai.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=[
        {
            "role": "user",
            "content": "Could you please help me identify a pain-point in 'Supply-Chain & Logistics' industry - something challenging that might be ripe for an Agentic solution?",
        }
    ],
)
pain_point = response.choices[0].message.content
display(Markdown(pain_point))
```
